2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
response_yes= df[df["Response"]== "Yes"]
```

```python
df_x= response_yes.groupby(["Policy Type", "Gender"])["Total Claim Amount"].mean().reset_index() 
df_x
```

```python
#The highest claim amount for female comes from special autos and for males it comes from personal autos
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
number_customers=df.groupby(["Policy Type", "State"])["Customer"].count().reset_index() 
number_customers
```

```python
countries_more_500= df.groupby("State")["Customer"].count().reset_index()
countries_more_500_x= countries_more_500[countries_more_500["Customer"]>500]
countries_more_500_x
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
df.groupby(["Gender", "Education"])["Customer Lifetime Value"].agg(["median", "max", "min"]).reset_index() 
```
